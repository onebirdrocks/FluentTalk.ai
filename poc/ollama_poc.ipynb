```python
ollama.embeddings(model='llama3', prompt='How to improve my English')
```

```python
response = ollama.chat(model='llama3', messages=[
  {
    'role': 'user',
    'content': 'How to improve my English?',
  },
])
print(response['message']['content'])
```

```python
stream = ollama.chat(
    model='llama3',
    messages=[{'role': 'user', 'content': 'How to improve my English?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
